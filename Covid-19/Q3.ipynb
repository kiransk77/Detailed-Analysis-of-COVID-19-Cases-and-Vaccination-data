# 3.For every district i, find the number of cases from the Covid-19 portal. Take the time-period of analysis from 15th March, 2020 to 14th August, 2021.

```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta 
```

```python
dist = "districts.csv"
cowin_data = 'cowin_vaccine_data_districtwise.csv' #distkey
```

```python
week = './Output_Files/Q3/cases-week.csv'
month = './Output_Files/Q3/cases-month.csv'
overall = './Output_Files/Q3/cases-overall.csv'
```

```python
#District_Key is taken from vaccine data and only req columns are taken into dataframe
vac_data = pd.read_csv(cowin_data,usecols = ['State','District','District_Key'],index_col=False)
covid_cases = pd.read_csv(dist,parse_dates=['Date'],usecols=['State','District','Date','Confirmed'],index_col=False)
```

```python
covid_cases = covid_cases[covid_cases['Date'] <= '2021-08-14']
```

```python
#Group by 'Date' column and give numbers started from 1 to each date, in newly created 'Dateid'
covid_cases['Dateid'] = covid_cases.groupby(['Date']).ngroup() + 1 
```

```python
#Creating month ids using newly created Date1 column
covid_cases['Date1'] = covid_cases["Date"] + timedelta(days=-14)
covid_cases['monthid'] = pd.DatetimeIndex(covid_cases['Date1']).month + 100*pd.DatetimeIndex(covid_cases['Date1']).year
covid_cases
```

```python
#5th Jan 2020, first Sunday of the year
covid_cases['Diff']=pd.to_datetime(covid_cases['Date'])-pd.to_datetime('2020-01-05')
covid_cases['dayid'] = ((covid_cases['Diff']/ np.timedelta64(1, 'D'))/7 ).astype(int)
covid_cases
```

```python
covid_cases['weekid'] = covid_cases.groupby(['dayid']).ngroup() + 1
covid_cases
```

```python
#df = df[df['monthid'] < 202109]
```

```python
covid_cases['Monthid'] = covid_cases.groupby(['monthid']).ngroup() + 1 
covid_cases
```

```python
covid_cases.drop(columns={"Diff"})
```

```python
data = pd.merge(vac_data,covid_cases,how='inner',left_on=['District','State'],right_on = ['District','State'])
data.rename(columns={'districtid':'District_Key'},inplace=True)
data
```

```python
check1=data[data["District_Key"]=="WB_Uttar Dinajpur"]
check1=check1[check1["Monthid"]==15]
check1
```

```python
data_month = data[['District_Key','Monthid','Confirmed']]
data_week = data[['District_Key','weekid','Confirmed']]
data_overall = data[['District_Key','Confirmed']]
```

```python
data_month = data_month.groupby(['Monthid','District_Key'])['Confirmed'].max().reset_index(name='Confirmed') 
data_week = data_week.groupby(['weekid','District_Key'])['Confirmed'].max().reset_index(name='Confirmed') 
data_overall = data_overall.groupby(['District_Key'])['Confirmed'].max().reset_index(name='Cases') 
```

```python
data_week['Cases'] = data_week.groupby(['District_Key'])['Confirmed'].diff().fillna(data_week['Confirmed'])
```

```python
data_month['Cases'] = data_month.groupby(['District_Key'])['Confirmed'].diff().fillna(data_month['Confirmed'])
```

```python
data_month[data_month["District_Key"]=="AP_East Godavari"]
```

```python
data_week = data_week[['District_Key','weekid','Cases']]
data_week.rename(columns={'District_Key':'districtid'},inplace=True)
```

```python
data_month = data_month[['District_Key','Monthid','Cases']]
data_month.rename(columns={'District_Key':'districtid','Monthid':'monthid'},inplace=True)
```

```python
data_overall.rename(columns={'District_Key':'districtid','Cases':'cases'},inplace=True)
```

```python
data_week.sort_values(["districtid","weekid"],inplace=True)
data_month.sort_values(["districtid","monthid"],inplace=True)
data_overall.sort_values("districtid",inplace=True)
```

```python
data_month.to_csv(month,index=False)
```

```python
data_week.to_csv(week,index=False)
```

```python
data_overall.to_csv(overall,index=False)
```
