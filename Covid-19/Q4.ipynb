```python
covid_cases = pd.read_csv(dist,parse_dates=['Date'],usecols=['State','District','Date','Confirmed'],index_col=False)
```

```python
vac_data1 = pd.read_csv(cowin_data,usecols=['State_Code','State'],index_col=False).drop_duplicates(subset=None, keep='first', inplace=False)
covid_cases = covid_cases[covid_cases['Date'] <= '2021-08-14']
```

```python
vac_data1.dropna(inplace=True)
```

```python
covid_cases['Date1'] = covid_cases["Date"] + timedelta(days=-14)
covid_cases['dateid'] = covid_cases.groupby(['Date1']).ngroup()+1
covid_cases['monthid'] = pd.DatetimeIndex(covid_cases['Date1']).month + 100*pd.DatetimeIndex(covid_cases['Date1']).year
```

```python
covid_cases
```

```python
#5th Jan 2020, first Sunday of the year
covid_cases['Diff1']=pd.to_datetime(covid_cases['Date'])-pd.to_datetime('2020-01-05')
covid_cases['dayid1'] = ((covid_cases['Diff1']/ np.timedelta64(1, 'D'))/7 ).astype(int)
```

```python
#5th Jan 2020, first Sunday of the year
covid_cases['Diff2']=(pd.to_datetime(covid_cases['Date'])-pd.to_datetime('2020-01-05'))/np.timedelta64(1, 'D')-4
covid_cases['dayid2'] = (covid_cases['Diff2']/7 ).astype(int)
```

```python
covid_cases
```

```python
covid_cases['weekid_1'] = covid_cases.groupby(['dayid1']).ngroup() + 1

covid_cases['weekid_2'] = covid_cases.groupby(['dayid2']).ngroup() + 1

covid_cases['monthid'] = covid_cases.groupby(['monthid']).ngroup() + 1 
```
