# 5.Find the number of people vaccinated with 1 or 2 doses of any vaccine, and sort the output file with district id and state id. Output this for all districts and all states weekly, monthly and overall in the following manner: districtid, timeid, dose1, dose2.

```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta 
```

```python
cowin_data = 'cowin_vaccine_data_districtwise.csv'
dist = "districts.csv"
```

```python
cols = pd.read_csv(cowin_data, nrows=1).columns
cols_d1=cols[9: :10]
cols_d1=["District_Key"] + cols_d1.tolist()
dist_data_dose_1 = pd.read_csv(cowin_data,usecols=cols_d1,index_col=False)
```

```python
cols_d2=cols[10: :10]
cols_d2=["District_Key"] + cols_d2.tolist()
dist_data_dose_2= pd.read_csv(cowin_data,usecols=cols_d2,index_col=False)
```

```python
dist_data_dose_1.dropna(subset=["District_Key"],inplace=True)
dist_data_dose_2.dropna(subset=["District_Key"],inplace=True)
```

```python
print(len(dist_data_dose_1["District_Key"]))
print(len(dist_data_dose_1["District_Key"].unique()))
#print(dist_data_dose_1[dist_data_dose_1["District_Key"]=="WB_South 24 Parganas"])
dist_data_dose_1= dist_data_dose_1.groupby(["District_Key"]).sum().reset_index()
#print(dist_data_dose_1[dist_data_dose_1["District_Key"]=="WB_South 24 Parganas"])
print(len(dist_data_dose_1["District_Key"]))
dist_data_dose_2= dist_data_dose_2.groupby(["District_Key"]).sum().reset_index()
print(len(dist_data_dose_2["District_Key"]))
```

```python
dist_data_dose_1 = pd.melt(dist_data_dose_1,id_vars=['District_Key'],
        var_name='Date', value_name='Cumulative_Dose1')
dist_data_dose_2 = pd.melt(dist_data_dose_2,id_vars=['District_Key'],
        var_name='Date', value_name='Cumulative_Dose2')
```

vac_data = pd.merge(dist_data_dose_1,dist_data_dose_2,how="inner",on="District_Key")

```python
dist_data_dose_1["Date"] = dist_data_dose_1.Date.str.split('.').str[0]
dist_data_dose_2["Date"] = dist_data_dose_2.Date.str.split('.').str[0]
```

```python
vac_data=pd.merge(dist_data_dose_1,dist_data_dose_2,how="inner",on=["District_Key","Date"])
```

```python
vac_data
```

```python
vac_data['dateid'] = vac_data.groupby(['Date']).ngroup()+1
#vac_data = vac_data[vac_data['Date'] <= '14/08/2021']
vac_data["Date"]=pd.to_datetime(vac_data.Date)


vac_data['Date1'] = vac_data["Date"] + timedelta(days=-14)
vac_data['monthid'] = pd.DatetimeIndex(vac_data['Date1']).month + 100*pd.DatetimeIndex(vac_data['Date1']).year

vac_data['Diff']=pd.to_datetime(vac_data['Date'])-pd.to_datetime('2021-01-03')
vac_data['weekid'] = ((vac_data['Diff']/ np.timedelta64(1, 'D'))/7 ).astype(int)
vac_data['monthid'] = vac_data.groupby(['monthid']).ngroup() + 1 
vac_data['State_Code']=vac_data.District_Key.str.split('_').str[0]

vac_data_weekly =vac_data[["State_Code","District_Key","Cumulative_Dose1","Cumulative_Dose2","weekid"]]
vac_data_monthly= vac_data[["State_Code","District_Key","Cumulative_Dose1","Cumulative_Dose2","monthid"]]
vac_data
```

```python
vac_data_weekly
```

```python
#vac_data_monthly["Cumulative_Dose1"] =vac_data_monthly["Cumulative_Dose1"].astype(long)
#vac_data_monthly["Cumulative_Dose2"] =vac_data_monthly["Cumulative_Dose2"].astype(long)
#vac_data_weekly["Cumulative_Dose1"] =vac_data_weekly["Cumulative_Dose1"].astype(long)
#vac_data_weekly["Cumulative_Dose2"] =vac_data_weekly["Cumulative_Dose2"].astype(long)
```

```python
vac_data_monthly["Cumulative_Dose1"] =pd.to_numeric(vac_data_monthly["Cumulative_Dose1"])
vac_data_monthly["Cumulative_Dose2"] =pd.to_numeric(vac_data_monthly["Cumulative_Dose2"])
vac_data_weekly["Cumulative_Dose1"] =pd.to_numeric(vac_data_weekly["Cumulative_Dose1"])
vac_data_weekly["Cumulative_Dose2"] =pd.to_numeric(vac_data_weekly["Cumulative_Dose2"])

vac_data_monthly_max = vac_data_monthly.groupby(['State_Code','monthid','District_Key']).agg({'Cumulative_Dose1':'max','Cumulative_Dose2':'max'}).reset_index() 
vac_data_weekly_max = vac_data_weekly.groupby(['State_Code','weekid','District_Key']).agg({'Cumulative_Dose1':'max','Cumulative_Dose2':'max'}).reset_index() 

vac_data_weekly_max['dose1'] = vac_data_weekly_max.groupby(['District_Key'])['Cumulative_Dose1'].diff().fillna(vac_data_weekly_max['Cumulative_Dose1'])
vac_data_weekly_max['dose2'] = vac_data_weekly_max.groupby(['District_Key'])['Cumulative_Dose2'].diff().fillna(vac_data_weekly_max['Cumulative_Dose2'])

vac_data_monthly_max['dose1'] = vac_data_monthly_max.groupby(['District_Key'])['Cumulative_Dose1'].diff().fillna(vac_data_monthly_max['Cumulative_Dose1'])
vac_data_monthly_max['dose2'] = vac_data_monthly_max.groupby(['District_Key'])['Cumulative_Dose2'].diff().fillna(vac_data_monthly_max['Cumulative_Dose2'])
```

```python
vac_data_weekly_max["dose1"] =pd.to_numeric(vac_data_weekly_max["dose1"])
vac_data_weekly_max["dose2"] =pd.to_numeric(vac_data_weekly_max["dose2"])
vac_data_monthly_max["dose1"] =pd.to_numeric(vac_data_monthly_max["dose1"])
vac_data_monthly_max["dose2"] =pd.to_numeric(vac_data_monthly_max["dose2"])
```

```python
vac_data_weekly_max.rename(columns={'District_Key':'districtid'},inplace=True)
vac_data_weekly_max
```

```python
vac_data_monthly_max.rename(columns={'District_Key':'districtid'},inplace=True)
vac_data_monthly_max
```

```python
dist_vac_data_overall=vac_data_monthly_max[["districtid","dose1","dose2"]]
dist_vac_data_overall = dist_vac_data_overall.groupby("districtid").agg({'dose1':'sum','dose2':'sum'}).reset_index()
dist_vac_data_overall
```

```python
dist_vac_data_weekly_final = vac_data_weekly_max[["districtid","weekid","dose1","dose2"]]
dist_vac_data_monthly_final=vac_data_monthly_max[["districtid","monthid","dose1","dose2"]]
dist_vac_data_overall_final=dist_vac_data_overall[["districtid","dose1","dose2"]]
```

```python
state_vac_data_weekly=vac_data_weekly_max[["State_Code","weekid","dose1","dose2"]]
state_vac_data_monthly=vac_data_monthly_max[["State_Code","monthid","dose1","dose2"]]
state_vac_data_overall = vac_data_monthly_max[["State_Code","dose1","dose2"]]
```

```python
state_vac_data_weekly_final=state_vac_data_weekly.groupby(["State_Code","weekid"]).agg({'dose1':'sum','dose2':'sum'}).reset_index()
state_vac_data_monthly_final=state_vac_data_monthly.groupby(["State_Code","monthid"]).agg({'dose1':'sum','dose2':'sum'}).reset_index()
state_vac_data_overall_final=state_vac_data_overall.groupby(["State_Code"]).agg({'dose1':'sum','dose2':'sum'}).reset_index()
```

```python
dist_vac_data_weekly_final.sort_values(["districtid","weekid"],inplace=True)
dist_vac_data_monthly_final.sort_values(["districtid","monthid"],inplace=True)
dist_vac_data_overall_final.sort_values(["districtid"],inplace=True)

state_vac_data_weekly_final.sort_values(["State_Code","weekid"],inplace=True)
state_vac_data_monthly_final.sort_values(["State_Code","monthid"],inplace=True)
state_vac_data_overall_final.sort_values(["State_Code"],inplace=True)
```

```python
dist_vac_data_weekly_final.to_csv("./Output_Files/Q5/district-vaccinated-count-week.csv",index=False)
dist_vac_data_monthly_final.to_csv("./Output_Files/Q5/district-vaccinated-count-month.csv",index=False)
dist_vac_data_overall_final.to_csv("./Output_Files/Q5/district-vaccinated-count-overall.csv",index=False)

state_vac_data_weekly_final.to_csv("./Output_Files/Q5/state-vaccinated-count-week.csv",index=False)
state_vac_data_monthly_final.to_csv("./Output_Files/Q5/state-vaccinated-count-month.csv",index=False)
state_vac_data_overall_final.to_csv("./Output_Files/Q5/state-vaccinated-count-overall.csv",index=False)
```
