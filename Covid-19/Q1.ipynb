```python
#json_data is json loaded 
#json_dist has keys of json
json_cowin_keys= json_cowin.keys()
res = dict([(key, val) for key, val in json_data.items() if key in json_cowin_keys])
len(res)
```

```python
res1={}
for i in res:
    key=json_cowin[i]
    value=set()
    for j in res[i]:
        if j in json_cowin_keys:
            jcvalue =json_cowin[j]
            if jcvalue.split("_")[0] != "DL":
                value.add(jcvalue)
            else:
                value.add("DL_Delhi")
    res1.update({key:list(value)})
```

```python
delhi=[]
neigh_delhi=set()
for i in res1:
    if i.split("_")[0]== "DL":
        delhi.append(i)
        for j in res1[i]:
            if j.split("_")[0] !="DL":
                neigh_delhi.add(j)
ndelhi=list(neigh_delhi)
modified_json = dict([(key, val) for key, val in res1.items() if key not in delhi])
modified_json.update({"DL_Delhi":ndelhi})
```

```python
from collections import OrderedDict
```

```python
dict1 = OrderedDict(sorted(modified_json.items()))
```

```python
finaldict={}
#SORTING ON DICT-VALUE LISTS
for key , value in dict1.items():
    finaldict[key] = sorted(value)
```

```python
with open('./Output_Files/Q1/neighbor-districts-modified.json','w') as q:
    json.dump(finaldict,q,indent=5)
```
